```python
import re
import nltk
nltk.download('stopwords')
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

lemmatizer = nltk.stem.WordNetLemmatizer()
from nltk.corpus import stopwords
stopwords.words('english')

```

```python
# Please note that the above code is quite easy to read:
# but it actually loops through the text three times,
# once for each line. We can do better:

def tknize (a_string):
    # Handle all the string operations at one time
    clean = re.sub('[^a-zA-Z \.]', ' ', a_string).lower()
    # Create the list of sub-strings (tokens) and return it
    clean_string = ''.join(clean)
    return nltk.tokenize.word_tokenize(clean_string)
    
```

```python
with open("1Doom.txt", mode ='r', encoding = 'UTF-8') as f:
    Doom = f.read()
print(Doom)
```

```python
nltk.download('punkt')
Doom_ = tknize(Doom)
Doom_
```

```python
nltk.download('averaged_perceptron_tagger')

Doom_tagged = nltk.pos_tag(Doom_)
Doom_tagged[15:35]
```

```python
reassembled = [tagged[0] for tagged in Doom_tagged]
print(" ".join(reassembled))
```

```python
for t in Doom_tagged:
    if t[1] == 'JJ':
        print(t[0])
```

```python
grammar = "Np: {<DT>?<JJ>*<NN>}"
parser = nltk.RegexpParser(grammar)
tree = parser.parse(Doom_tagged)
for i in tree[0:20]:
    print(i)
```
