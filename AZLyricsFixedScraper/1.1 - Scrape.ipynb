```python
#Imports block
#Import urllib.request to get a validated connection
import urllib.request
#Import time to calculate remaining time for scraping
import time
#Import random to generate random numbers to throw off robots.txt
import random
#Import to find the specific div that contains our information
from bs4 import BeautifulSoup
#Import to modify files in the system
import os
```

```python
# URL to the webpage for AZlyrics. Because this is a specific case, the 
# url was fixed(Hence fixed scraper). To change that, I would add an input
# statement asking for the AZLyrics artist's link 
# (All AZLyrics pages have the same structure)
myurl = "https://www.azlyrics.com/m/mfdoom.html"

# Initial connection to the url to ensure user wont be blocked
myconnection = urllib.request.urlopen(myurl)
myhtml = myconnection.read()
# Print successful connection to website
print(myhtml[0:100])
```

```python
# Go through the website html and grab all song titles from artist page
mysoup = BeautifulSoup(myhtml, "html")
# Finds all divs with "listalbum-item"
mydivs = mysoup.find_all(class_="listalbum-item")
print(type(mydivs))
count = 0
# This will find all song titles
# Includes album names listed under class = "comment"
for i in mydivs:
    count+=1
    print(f"({count}) {i}")
```

```python
#This block essentially prepares the links in a payload for the urllib.request
links = []
# The block here is due to the kernel breaking after 140 songs
# Wasn't sure how to fix it. I also used this block for testing
# 10 songs at a time. This will change the amount of songs you will scrape
# Finally, putting a value above the actual max won't throw an error
maxSongs = int(input("Enter maximum number of songs to find: "))
# Counter will count each song added. This will 
counter = 0
# Iterate over the list of divs of the website to the max number of items
# If the div contains the class "comment", then it is ignored as it's an album, not
# a song. The counter will also validate how many songs are truly added to the list
for div in mydivs[0:maxSongs]:
    if not div.find('div',class_= 'comment'):
        the_link = div.find('a')['href']
        links.append(the_link)
        counter+=1
    print(f"({counter}) {div}")
# Check results of links variable
print(links)
print(f"Number of songs appended to list: {counter}")
```

```python
#Go to base website of azlyrics.com
baseURL = "https://www.azlyrics.com"
folderName = input("Enter a name for the artists folder: ")

# When we go to save the files, 
# we want to remove this part of the link string:
myfilter = "/lyrics/mfdoom/"
songTrack = 0
timeLeft = (len(links) * 30)
for link in links:
    remotefile = urllib.request.urlopen(baseURL+link)
    filename = link.replace(myfilter, '')
    #-----Sleep Function to throw off robots.txt
    time.sleep(random.randint(20,30))
    #-----If you remove you will get IP Banned
    
    localfile = open(filename,'wb')
    localfile.write(remotefile.read())
    localfile.close()
    remotefile.close()
    print(f"({songTrack}) Minutes Remaining: {timeLeft/60} {link}")
    timeLeft-=30
    songTrack+=1
```
